```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# fmri_img_list : NiftiImage object, fmri_data_list: np array object
fmri_img_list = []
fmri_data_list = []
for i in range(test_num):
  fmri_img_list.append(nib.load(f'/content/data/derivatives/sub-CN00{i+1}/func/sub-CN00{i+1}_task-lppCN_run-{sub_run_list[i]}_space-MNIColin27_desc-preproc_bold.nii'))
  fmri_data_list.append(fmri_img_list[i].get_fdata()) 
```

```python
print(fmri_img_list[0])
```

So this is a `Nifti1Image` object which contains 3 things :
* some data of shape (73, 90, 74, 283)
* an affine array that defines the spatial orientation and scale of the data
* a header, containing more information about the data format

Note that the data is a 4D array, and from the 5th element of pixdim we could know that the data was record every 2 second, which means 566s scanning (which matches the data in annotation\CN\lppCN_section1.TextGrid )

### Merge 9 participants' data into single image

```python
# Merge all image
import numpy as np

all_data = 0
# Sum the FMRI images
for image_data in fmri_data_list:
    all_data += image_data

# Take the average of the summed images
all_data /= len(fmri_data_list)

# Create a new NIfTI image using the first image as a template
merged_image = nib.Nifti1Image(all_data,affine=fmri_img_list[0].affine)

# Save the merged image to a file
nib.save(merged_image, "merged_image_CH.nii")
```
